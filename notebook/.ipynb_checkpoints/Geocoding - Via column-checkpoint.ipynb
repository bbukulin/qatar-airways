```python
import pandas as pd
from geopy.geocoders import Nominatim
import time
```

```python
data = pd.read_csv('cleaned_qatar_airways_reviews.csv')
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
def geocode_location(location):
    try:
        return geolocator.geocode(location)
    except:
        return None
```

```python
# Geokodiranje jedinstvenih lokacija
# Geokodiranje jedinstvenih lokacija
unique_locations = set(data['From'].tolist() + data['To'].tolist())
if 'Direct' not in unique_locations:
    unique_locations.update(data[data['Via'] != 'Direct']['Via'].tolist())
location_coords = {}

for location in unique_locations:
    geocoded = geocode_location(location)
    if geocoded:
        location_coords[location] = (geocoded.latitude, geocoded.longitude)
    time.sleep(1)  # Pauza kako bi se izbjegao preveliki broj zahtjeva u kratkom vremenu
```

```python
# Dodavanje koordinata u originalne podatke
data['FromLat'] = data['From'].apply(lambda x: location_coords[x][0] if x in location_coords else None)
data['FromLon'] = data['From'].apply(lambda x: location_coords[x][1] if x in location_coords else None)
data['ToLat'] = data['To'].apply(lambda x: location_coords[x][0] if x in location_coords else None)
data['ToLon'] = data['To'].apply(lambda x: location_coords[x][1] if x in location_coords else None)
data['ViaLat'] = data['Via'].apply(lambda x: location_coords[x][0] if x in location_coords and x != 'Direct' else None)
data['ViaLon'] = data['Via'].apply(lambda x: location_coords[x][1] if x in location_coords and x != 'Direct' else None)
```

```python
data.to_csv('cleaned_qatar_airways_reviews_with_coords_corrected.csv', index=False)
```
